```python
import os
import pandas as pd
from tqdm import tqdm

directory = "../data/challenge-2019/physionet.org/files/challenge-2019/1.0.0/training"
data = {}

def patient_files():
    for root, dirs, files in os.walk(directory):
        for filename in files:
            if filename.endswith(".psv"):
                yield os.path.join(root, filename)

for filepath in tqdm(patient_files(), total=40_336):
    patient_sample = pd.read_csv(filepath, delimiter="|")
    # Extract the file name without extension as the id
    patient_id = os.path.splitext(os.path.basename(filepath))[0]
    # Add the data to the table with the id as the value
    data[int(patient_id[1:])] = patient_sample

data[1]
```

```python
table_lengths = [len(table) - 1 for table in data.values()]
min_rows = min(table_lengths)
max_rows = max(table_lengths)
avg_rows = sum(table_lengths) / len(table_lengths)

print("Minimum number of rows:", min_rows)
print("Maximum number of rows:", max_rows)
print("Average number of rows:", avg_rows)
```

```python
processed_data = {}

window_size = 7
processed_data = {}

for key, value in tqdm(data.items(), desc="Transforming tables"):
    table = value.copy()
    num_rows = len(table)
    num_windows = num_rows - window_size + 1
    
    for i in range(num_windows):
        window = table.iloc[i:i+window_size]
        window_key = (key, i)
        processed_data[window_key] = window

for key, value in tqdm(processed_data.items(), desc="Saving data on disk"):
    filepath = f"../data/challenge-2019/processed_samples/{key[0]}_{key[1]}.csv"
    value.to_csv(filepath, index=False)
    
print("Number of sequences for training:", len(processed_data))
```
